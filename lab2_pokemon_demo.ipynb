# Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Importing Data

```python
df = pd.read_csv('d:\DAP391m\Pokemon\pokemon-1.csv')
df.head()
```

```python
# Clean column names (remove extra spaces)
df.columns = df.columns.str.strip()
```

```python
stats_summary = df.describe()
stats_summary
```

```python
# Plot 1: Khối lượng và chiều cao của Pokémon 
plt.figure(figsize=(10, 6))
plt.hist(df['weight_kg'].dropna(), bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution of Pokémon Weight')
plt.xlabel('Weight (kg)')
plt.ylabel('Number of Pokémon')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Plot 2: Phân tích thống kê về chỉ số Pokémon.
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='attack', y='defense', hue='generation', palette='viridis', alpha=0.7)
plt.title('Attack vs Defense by Generation')
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.legend(title='Generation')
plt.show()
```

```python
# Plot 3: Hiệu quả của loại Pokémon và Phân tích thống kê về chỉ số Pokémon.
plt.figure(figsize=(14, 7))
sns.boxplot(data=df, x='type1', y='speed', palette='pastel')
plt.title('Speed Distribution by Primary Type')
plt.xlabel('Primary Type')
plt.ylabel('Speed')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot 4: Phân tích thống kê về chỉ số Pokémon và Hiệu quả của các chỉ số đặc biệt (SP Attack, SP Defense).
plt.figure(figsize=(12, 10))
correlation_matrix = df[['hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Stats')
plt.show()
```

```python
# Plot 5: Khác biệt giữa Pokémon huyền thoại và không huyền thoại và Phân tích đặc tính "Legendary" và "Generation"
legendary_counts = df[df['is_legendary'] == 1]['generation'].value_counts().sort_index()
legendary_counts.plot(kind='bar', color='gold', figsize=(10, 6))
plt.title('Count of Legendary Pokémon by Generation')
plt.xlabel('Generation')
plt.ylabel('Number of Legendary Pokémon')
plt.xticks(rotation=0)
plt.show()
```

```python
# Plot 6: Phân tích thống kê về chỉ số Pokémon và Hiệu quả của các chỉ số đặc biệt (SP Attack, SP Defense).
plt.figure(figsize=(10, 6))
plt.hist(df['attack'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Pokémon Attack Stats')
plt.xlabel('Attack Stat')
plt.ylabel('Number of Pokémon')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Plot 7: Phân bố giới tính của Pokémon và mối quan hệ với chỉ số
# Gender distribution
gender_distribution = df['percentage_male'].value_counts(bins=[0, 50, 100]).sort_index()

# Plot pie chart
plt.figure(figsize=(8, 8))
gender_distribution.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Gender Distribution of Pokémon")
plt.ylabel("")
plt.show()

# Compare stats by gender
plt.figure(figsize=(12, 6))
sns.boxplot(x='percentage_male', y='base_total', data=df)
plt.title("Base Total by Gender Distribution")
plt.xlabel("Percentage Male")
plt.ylabel("Base Total")
plt.show()
```

```python
# Plot 8 : Sự khác biệt giữa Pokémon hệ kép (Dual-type) và hệ đơn (Single-type)
# Create a column to classify Pokémon by type count
df['type_count'] = df[['type1', 'type2']].notnull().sum(axis=1)

# Calculate average stats for each type count
avg_stats = df.groupby('type_count')['base_total'].mean()

# Plot comparative bar chart
plt.figure(figsize=(8, 6))
avg_stats.plot(kind='bar')
plt.title("Average Base Total: Single-type vs Dual-type Pokémon")
plt.xlabel("Type Count (1 = Single-type, 2 = Dual-type)")
plt.ylabel("Average Base Total")
plt.xticks(rotation=0)
plt.show()
```

```python
# Plot 9 : Xu hướng phát hành Pokémon qua các thế hệ
# Count Pokémon by generation
generation_counts = df['generation'].value_counts().sort_index()

# Plot line chart
plt.figure(figsize=(10, 6))
generation_counts.plot(kind='line', marker='o')
plt.title("Number of Pokémon Released by Generation")
plt.xlabel("Generation")
plt.ylabel("Number of Pokémon")
plt.xticks(generation_counts.index)
plt.grid()
plt.show()
```

```python
# Plot 10 : Tần suất xuất hiện Pokémon huyền thoại trong các trận đấu cạnh tranh
# Count the number of legendary Pokémon
legendary_counts = df[df['is_legendary'] == 1].shape[0]

# Count the number of normal Pokémon
normal_counts = df[df['is_legendary'] == 0].shape[0]

# Plot bar chart
plt.figure(figsize=(8, 6))
plt.bar(['Legendary', 'Normal'], [legendary_counts, normal_counts])
plt.title("Legendary vs Normal Pokémon Usage in Competitive Battles")
plt.ylabel("Number of Pokémon")
plt.show()
```

```python
# Plot 11: Tính cân bằng giữa các thế hệ Pokémon (Power Creep)
# Calculate average total stats by generation
avg_stats_by_gen = df.groupby('generation')['base_total'].mean()
plt.figure(figsize=(12, 6))
avg_stats_by_gen.plot(kind='line', marker='o')
plt.title('Average Total Stats Across Pokémon Generations')
plt.ylabel('Average Total Stats')
plt.xlabel('Generation')
plt.xticks(range(1, df['generation'].max() + 1))
plt.grid()
plt.show()
```

```python
# Plot 12 : Hiệu quả của Pokémon theo từng môi trường hoặc vùng miền
# Calculate average stats by type
type_stats = df.groupby('type1')['base_total'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
type_stats.plot(kind='bar')
plt.title('Average Total Stats of Pokémon Types')
plt.ylabel('Average Total Stats')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot 13 : Phân tích sự tương quan giữa loại (Type) và chỉ số đặc biệt (SP Attack, SP Defense)
# Create a heatmap for Special Attack and Special Defense by type
heatmap_data = df.groupby('type1')[['sp_attack', 'sp_defense']].mean().reset_index()
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data.set_index('type1'), annot=True, cmap='coolwarm')
plt.title('Correlation Between Pokémon Type and Special Stats')
plt.xlabel('Stat')
plt.ylabel('Pokémon Type')
plt.show()
```

```python
# Plot 15 : Phân tích top 10 Pokémon có khả năng phổ biến nhất 
# Clean the 'abilities' column (remove square brackets and quotes)
df['abilities'] = df['abilities'].str.strip("[]").str.replace("'", "").str.split(',')

# Count unique abilities
abilities = pd.Series([ability.strip() for sublist in df['abilities'].dropna() for ability in sublist])

# Plot the top 10 abilities
plt.figure(figsize=(10, 6))
sns.barplot(y=abilities.value_counts().head(10).index, x=abilities.value_counts().head(10))
plt.title("Top 10 Most Common Pokémon Abilities")
plt.xlabel("Number of Pokémon")
plt.ylabel("Ability")
plt.show()



```

```python
# Plot 15 : Hiệu quả của các chỉ số đặc biệt (SP Attack, SP Defense) so với chỉ số vật lý
# Create a scatter plot to compare SP Attack, SP Defense, and Physical Attack
plt.figure(figsize=(10, 6))

# Scatter plot between SP Attack and SP Defense
sns.scatterplot(x=df['sp_attack'], y=df['sp_defense'], hue=df['type1'])
plt.title('Special Attack vs Special Defense with Pokémon Types')
plt.xlabel('Special Attack')
plt.ylabel('Special Defense')
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
plt.show()

# Compare Physical Attack and Defense with SP Attack and SP Defense
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['attack'], y=df['defense'], hue=df['type1'])
plt.title('Physical Attack vs Defense with Pokémon Types')
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
plt.show()
```
